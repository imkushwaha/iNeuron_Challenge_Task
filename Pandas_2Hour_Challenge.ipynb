### Remove  dollar sign from item_price and make it numeric to calculate:
#### total_ordered_value = item_price * quantity

```python
len_item_price = len(chipotle_data["item_price"])
```

```python
len_item_price
```

```python
for i in range(0,len_item_price):
    chipotle_data["item_price"][i] = float(chipotle_data["item_price"][i].split("$")[1])
```

```python
chipotle_data.head()
```

```python
#Adding new column with name total_ordered_value 

chipotle_data["total_ordered_value"] = chipotle_data["quantity"]*chipotle_data["item_price"]
```

```python
chipotle_data.head()
```

```python
#maximum ordered value
max(chipotle_data["total_ordered_value"])
```

### Details of maximum ordered value

```python
chipotle_data["total_ordered_value"].sort_values(ascending = False)
```
