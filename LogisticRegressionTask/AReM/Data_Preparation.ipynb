### Applying same code for whole walking folder

```python
try:
    #empty data frame with same column
    walking_df = pd.DataFrame(columns = ["time", "avg_rss12", "var_rss12", "avg_rss13", "var_rss13", "avg_rss23", "var_rss23", "label"])
    
    for i in os.listdir("walking"):
        data = open(f"walking//{i}", "rb").readlines()
        main_list = []

        for i in data[5:]:
            l1 = str([i]).split(",")
            main_list.append(l1)
        
        df = pd.DataFrame(main_list)
    
        for i in range(len(df)):
            df[0][i] = int(df[0][i].split("[b'")[1])
            
        for i in range(len(df)):
            df[6][i] = float(df[6][i].split("\\n")[0])
    
        df["label"] = "walking"
    
        col_names = ["time", "avg_rss12", "var_rss12", "avg_rss13", "var_rss13", "avg_rss23", "var_rss23", "label"]
    
        df.columns = col_names
        
        walking_df = pd.concat([df,walking_df])
        
    #reseting index     
    walking_df = walking_df.reset_index()
    
    #droping index column
    walking_df.drop("index", inplace = True, axis = 1)
    
except Exception as e:
    print("Error occured",e)
```

```python
walking_df
```
