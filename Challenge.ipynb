# Spliitig Data set to make it in desire format

- First splitting by ","
- And then splitting by ";"
- Finally we making list of list with each list having 14 elements

```python
import csv

l = []
final_list = []

with open("carbon_nanotubes.csv", "r") as data :
    next(data)
    data_csv = csv.reader(data, delimiter = "\n")

    
    for i in (data_csv):
        for j in i:
            split_1 = j.split(",")       
        
        for j in split_1:
            split_2 = j.split(";")
            l.append(split_2)
            
    
    for i in l:
        for j in i:
            final_list.append(j)
            
    my_list = final_list
    n = 14 #number of columns
    data = [my_list[i:i + n] for i in range(0, len(my_list), n)]
```
