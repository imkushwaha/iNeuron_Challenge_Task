# Loading required libraries

```python
import logging as lg
import os
import csv
import mysql.connector as connection
```

## Checking if file present in working directory or not

```python
for i in os.listdir():
        if i == "carbon_nanotubes.csv":
            print("carbo_nanotues.csv present in your working directory !!")
```

##  Appending all single list into a main list

```python
try:
    
    main_list = []

    with open("carbon_nanotubes.csv", "r") as data :
        next(data)
        data_csv = csv.reader(data, delimiter = ";")
        print(data_csv)
        for i in (data_csv):
            main_list.append(i)
        
            
except Exception as e:
    print(f"Your error is: {e}")
    
    
else:
    print("Code executed successfully !!")
```

## Creating new csv file of main_list with each line under single quote 

```python
with open('New_Carbon_nanotubes.csv', "w", newline = '') as df:
    writer = csv.writer(df)
    writer.writerows(main_list)
```

## Making Connection to db and will create table with 8 columns 

```python
try:
    lg.basicConfig(filename = "sql_task.log", level = lg.INFO, format = '%(asctime)s %(message)s',datefmt='%Y-%m-%d %H:%M:%S')
    
    H = str(input("Please enter your host id: "))
    U = str(input("Please enter your user name: "))
    P = str(input("Please enter your Password: "))
    
    lg.info(f"Host id {H}, User name {U} has tried to make a connection !!")
    
    mydb = connection.connect(host=H, user=U, passwd=P,use_pure=True)
    #check if the connection is established
    print(mydb.is_connected())
    lg.info("Connection successfull !!")
    
    cur = mydb.cursor()

    query = "CREATE TABLE student.carbon_data (Chiral_indice_n INT(10),Chiral_indice_m INT(10), Initial_atomic_coordinate_u VARCHAR(20), Initial_atomic_coordinate_v VARCHAR(20), Initial_atomic_coordinate_w VARCHAR(20)," \
           "Calculated_atomic_coordinates_u VARCHAR(20), Calculated_atomic_coordinates_v VARCHAR(20), Calculated_atomic_coordinates_w VARCHAR(20))"

    cur.execute(query)
    lg.info("One table created !!")
    
except Exception as e:
    print("you can check your log for more info if your code will fail")
    lg.error("error has occured")
    lg.exception(str(e))
    
else:
    print("code executed successfully !!")
    
```

## Inserting data to created table carbon_data

```python
try:
    
    with open("New_Carbon_nanotubes.csv", "r") as data :
        next(data)
        data_csv = csv.reader(data, delimiter = "\n")
        print(data_csv)
        lg.info("Data read successfully")
        for i in enumerate(data_csv):
            #print(i)
            for j in i[1]:
                cur.execute('insert into student.carbon_data values ({data})'.format(data = j))
                
        lg.info("All data inserted successfully")  
        
except Exception as e:
    print("you can check your log for more info if your code will fail")
    lg.error("error has occured")
    lg.exception(str(e))
    
else:
    print("all the data inserted") 
            
finally :
    mydb.commit()
```
